```python
import re
import pandas as pd
```

```python
# Prepare data
trump = pd.read_csv('tweet_data.csv')
```

```python
trump.Tweet = trump.apply(lambda row: re.sub(r"http\S+", "", row.Tweet).lower(), 1)
trump.Tweet = trump.apply(lambda row: " ".join(filter(lambda x:x[0]!="@", row.Tweet.split())), 1)
trump.Tweet = trump.apply(lambda row: " ".join(re.sub("[^a-zA-Z]+", " ", row.Tweet).split()), 1)
timestamps = trump.Date.to_list()
tweets = trump.Tweet.to_list()
```
